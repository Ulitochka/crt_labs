```python
actual = pd.read_pickle('actual.pckl')
predictions = pd.read_pickle('predcitions.pckl')

file_paths = actual['head']
arousal = actual['head.2']
valence = actual['head.1']
```

```python
XA_train = []
XV_train = []
predict = np.zeros((0, 2), dtype=np.float32)
for elem in tqdm_notebook(file_paths):
    data = torch.randn(96, 96, dtype=torch.float32) #data 
    logits = model(data).reshape(1,2)
    predict = np.concatenate((predict, logits.cpu().data.numpy()), axis=0)

XA_train = predict[:,1]
XA_train= XA_train.reshape(len(XA_train), 1)

XV_train = predict[:,0]
XV_train = XV_train.reshape(len(XV_train), 1)
```

```python
pred = pd.DataFrame(np.hstack((XV_train, XA_train)), columns = ['{}_variance'.format(model_name), '{}_arousal'.format(model_name)])
pred = pd.concat([file_paths, pred], axis = 1)
predictions.merge(pred, on = ['head'], inplace=True)

predictions.to_pickle('predcitions.pckl')
```
